```python
# 2. Write a program to calculate the factorial of a number provided by the user. 

num = int(input("Enter a number: "))
fact = 1
i = 1

while i <= num:
    fact = fact * i
    i = i + 1

print("Factorial of", num, "is", fact)

```

```python

# 6. Write a program to generate the Fibonacci series up to a number provided by the  user. 

limit = int(input("Enter the limit: "))

a, b = 0, 1

print("Fibonacci series up to", limit, ":")
while a <= limit:
    print(a, end=" ")
    a, b = b, a + b
```

```python
# 7. Write a program to check whether a number provided by the user is prime. 

num = int(input("Enter a number: "))
if num <= 1:
    print(num, "is not a Prime number")
else:
    i = 2
    is_prime = True
    while i <= num // 2:
        if num % i == 0:
            is_prime = False
            break
        i += 1
    
    if is_prime:
        print(num, "is a Prime number")
    else:
        print(num, "is not a Prime number")
```

```python
# 11. Write a program that takes the starting and ending points of a range as input 
# from the user, then calculates the sum of all numbers within that range. 
start = int(input("Enter the starting number: "))
end = int(input("Enter the ending number: "))

total = 0
i = start

while i <= end:
    total += i
    i += 1

print(f"The sum of numbers from {start} to {end} is:", total)
```

```python
# 12. Write a program that takes the starting and ending points of a range as input 
# from the user, then calculates the product of all numbers within that range. 
start = int(input("Enter the starting number: "))
end = int(input("Enter the ending number: "))

product = 1
i = start

while i <= end:
    product *= i
    i += 1

print(f"The product of numbers from {start} to {end} is:", product)
```

```python
#15. Write a program to filter out all vowels and consonants from a string provided by the user. 
text = input("Enter a string: ")
vowels = "aeiouAEIOU"
only_vowels = ""
only_consonants = ""

i = 0
while i < len(text):
    if text[i].isalpha():         
        if text[i] in vowels:    
            only_vowels += text[i]
        else:                     
            only_consonants += text[i]
    i += 1

print("Vowels in the string:", only_vowels)
print("Consonants in the string:", only_consonants)
```

```python
#16. Write a program to filter out duplicate characters from a string provided by the  user. 
text = input("Enter a string: ")
result = ""
i = 0

while i < len(text):
    if text[i] not in result:  
        result += text[i]
    i += 1

print("String without duplicates:", result)
```

```python
#18. Write a program to display all letters except 'm' and 'i' from the string "Dreamer Infotech". 
text = "Dreamer Infotech"
result = ""

for char in text:
    if char.lower() not in ['m', 'i']:  
        result += char
print("Resulting string:", result)
```

```python
#20. Write a program to reverse a string provided by the user. 

text = input("Enter a string: ")
reversed_text = ""
i = len(text) - 1   

while i >= 0:
    reversed_text += text[i]
    i -= 1
print("Reversed string:", reversed_text)
```
