```python
#  1.Sum of all keys: 
# Write a program to calculate the sum of all keys in the dictionary: 
# d = {1: 1, 2: 2, 3: 3, 4: 4} 
d = {1: 1, 2: 2, 3: 3, 4: 4} 
res=sum(d.keys())
print(res)
```

```python
# 2.Sum of all values: 
# Write a program to calculate the sum of all values in the dictionary: 
# d = {1: 1, 2: 2, 3: 3, 4: 4} 
d = {1: 1, 2: 2, 3: 3, 4: 4}
res=sum(d.values())
print(res)
```

```python
# 3.Sum of keys and values: 
# Write a program to calculate the sum of both keys and values in the 
# dictionary: 
# d = {1: 1, 2: 2, 3: 3, 4: 4} 
d = {1: 1, 2: 2, 3: 3, 4: 4} 
res = sum(d.keys()) + sum(d.values())
print(res)
```

```python
# 7.Find the greatest key: 
# Write a program to find the greatest key in the dictionary: 
# player = {7: "Dhoni", 12: "Kohli", 9: "Rohit", 89: "Bumrah"}

player = {7: "Dhoni", 12: "Kohli", 9: "Rohit", 89: "Bumrah"}
greatest_key = max(player.keys())
print("Greatest key:", greatest_key)
print("Value of greatest key:", player[greatest_key])
```

```python
# 8.Find alternate items: 
# Write a program to extract alternate items (key-value pairs) from the 
# dictionary: 
# player = {7: "Dhoni", 12: "Kohli", 9: "Rohit", 89: "Bumrah"}

player = {7: "Dhoni", 12: "Kohli", 9: "Rohit", 89: "Bumrah"}
items = list(player.items())      
alternate_items = dict(items[::2])

print("Alternate items:", alternate_items)
```

```python
# 11.Check if a key exists: 
# Write a program to check whether a given key exists in the 
# dictionary: 
# d = {1: 100, 2: 200, 3: 300} 

d = {1: 100, 2: 200, 3: 300}
key =int(input("enter a value"))
if key in d.keys():
    print(f"Key {key} exists with value {d[key]}")
else:
    print(f"Key {key} does not exist")
```

```python
# 14.Reverse key-value pairs: 
# Write a program to swap keys with values in a dictionary: 
# d = {1: "one", 2: "two", 3: "three"} 
d = {1: "one", 2: "two", 3: "three"} 
reversed_d = {x: k for k, x in d.items()}
print(reversed_d)
```

```python
# 19.Sort dictionary by keys: 
# Write a program to sort a dictionary by its keys. 
# d = {3: "three", 1: "one", 2: "two"}

d = {3: "three", 1: "one", 2: "two"}
sort_dict = dict(sorted(d.items()))
print("Sorted dictionary:", sort_dict)
```
