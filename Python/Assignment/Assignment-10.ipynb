```python
# 15. Credit Card Generation 
# Write a function generate_10_card(ac_holder) that: 
# ‚óè Uses the bank name ‚ÄúPUNJAB NATIONAL BANK‚Äù 
# ‚óè Generates a 16-digit card number (formatted with spaces every 4 digits) 
# ‚óè Sets Valid From as current month/year (MM/YY) 
# ‚óè Sets Expiry Date to 5 years from now 
# ‚óè Generates a random 3-digit CVV 
# ‚óè Saves all card details into card.txt 
# Sample Output: 
# Bank: PUNJAB NATIONAL BANK 
# Account Holder: Sandeep 
# Card Number: 0778 3034 7224 1724 
# Valid From: 09/24 
# Expiry Date: 09/29 
# CVV: 425 

import random, datetime

def generate_10_card(ac_holder):
    now = datetime.datetime.now()
    vf = now.strftime("%m/%y")
    exp = now.replace(year=now.year+5).strftime("%m/%y")
    with open("card.txt", "a") as f:
        for _ in range(10):
            s = ''.join(str(random.randint(0,9)) for _ in range(16))
            num = ' '.join(s[i:i+4] for i in range(0,16,4))
            cvv = random.randint(100,999)
            f.write(f"Bank: PUNJAB NATIONAL BANK\nAccount Holder: {ac_holder}\nCard Number: {num}\nValid From: {vf}\nExpiry Date: {exp}\nCVV: {cvv}\n\n")
    print("Saved 10 cards to card.txt")

# example
generate_10_card("Sandeep")

```

```python
# 20. Student Result Classification 
# Write a Python program to classify students into Pass and Fail based on marks. 
# If marks ‚â• 33 ‚Üí write to pass_students.txt 
# If marks < 33 ‚Üí write to fail_students.txt 
# Sample Data: 
# students = { 
# "Ravi": {"marks": 75, "class": "10A"}, 
# "Priya": {"marks": 29, "class": "10B"}, 
# "Aman": {"marks": 45, "class": "10A"}, 
# "Sita": {"marks": 18, "class": "10C"}, 
# "Karan": {"marks": 55, "class": "10B"}, 
# "Geeta": {"marks": 30, "class": "10A"}, 
# } '
# 20. Student Result Classification


students = {
    "Ravi": {"marks": 75, "class": "10A"},
    "Priya": {"marks": 29, "class": "10B"},
    "Aman": {"marks": 45, "class": "10A"},
    "Sita": {"marks": 18, "class": "10C"},
    "Karan": {"marks": 55, "class": "10B"},
    "Geeta": {"marks": 30, "class": "10A"},
}


with open("pass_students.txt", "w") as pass_file, open("fail_students.txt", "w") as fail_file:
    # Write headers
    pass_file.write("Name\tMarks\tClass\n")
    fail_file.write("Name\tMarks\tClass\n")

    # Classify students
    for name, details in students.items():
        marks = details["marks"]
        cls = details["class"]
        
        if marks >= 33:
            pass_file.write(f"{name}\t{marks}\t{cls}\n")
        else:
            fail_file.write(f"{name}\t{marks}\t{cls}\n")

print(" Student classification completed. Files generated:")
print(" pass_students.txt")
print(" fail_students.txt")
```

```python
# 47. Rename Files in Bulk 
# Write a Python program that renames all .txt files in a directory by adding today‚Äôs date as a 
# prefix.

import os
from datetime import datetime

def rename_txt_files(directory):
    date_prefix = datetime.now().strftime("%d-%m-%Y_")
    for file in os.listdir(directory):
        if file.endswith(".txt"):
            old_path = os.path.join(directory, file)
            new_path = os.path.join(directory, date_prefix + file)
            os.rename(old_path, new_path)
    print("All .txt files renamed successfully.")

# Example usage
rename_txt_files(".")  # current directory
```

```python
# 48. Display File Size and Creation Time 
# Write a program that displays the file name, file size (in KB), and creation date/time for all text 
# f
#  iles in a directory.
import os
from datetime import datetime

def file_info(directory):
    for file in os.listdir(directory):
        if file.endswith(".txt"):
            path = os.path.join(directory, file)
            size_kb = os.path.getsize(path) / 1024
            ctime = datetime.fromtimestamp(os.path.getctime(path)).strftime("%d-%m-%Y %H:%M:%S")
            print(f"File: {file}\nSize: {size_kb:.2f} KB\nCreated: {ctime}\n{'-'*40}")

# Example usage
file_info(".")  # Current directory
```

```python
# 49. Split a File Based on Word Count 
# Write a program that splits a large text file into smaller files, 
# each containing exactly 100 words (not lines).
def split_file_by_words(input_file):
    with open(input_file, "r") as f:
        words = f.read().split()
    
    for i in range(0, len(words), 100):
        part = words[i:i+100]
        with open(f"part_{i//100 + 1}.txt", "w") as f:
            f.write(' '.join(part))
    
    print("File successfully split into smaller files of 100 words each.")

# Example usage
split_file_by_words("myfile.txt")
```
