```python
# 4. Write a Python program to check if a number provided by the user is prime or not.
num = int(input("Enter a number: "))
if num <= 1:
    print(f"{num} is not a prime number.")
else:
    is_prime = True
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(f"{num} is a prime number.")
    else:
        print(f"{num} is not a prime number.")

```

```python
# 7. Write a Python program to generate the Fibonacci sequence up to a specified number of terms.
n = int(input("Enter the number: "))
a, b = 0, 1
for _ in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

```python
# 16. Write a Python program to filter out all vowels and consonants from a string 
# entered by the user.

text = input("Enter a string: ")
vowels = "aeiouAEIOU"
only_vowels = ""
only_consonants = ""

for char in text:              
    if char.isalpha():           
        if char in vowels:      
            only_vowels += char
        else:                   
            only_consonants += char

print("Vowels:", only_vowels)
print("Consonants:", only_consonants)
```

```python
# 20. Write a Python program to find duplicate letters between two strings. 
# Example: In "virat" and "rohit", the duplicate letter is "r". 


str1 = input("Enter first string: ")
str2 = input("Enter second string: ")
duplicates = ""
for char in str1:
    if char in str2 and char not in duplicates:   
        duplicates += char

if duplicates:
    print("Duplicate letters:", duplicates)
else:
    print("No duplicate letters found.")

```

```python
# 24. Given: text = "programming" 
# Goal: Print all characters that repeat in the string.
text = "programming"
repeats = ""
for i in text:           
    count = 0
    for j in text:        
        if i == j:
            count += 1
    if count > 1 and i not in repeats:
        repeats += i

print("Repeated characters:", repeats)
```

```python
# 26. Given: text = "knowyourself" 
# Goal: Find and print the first character that repeats.

text = "knowyourself"
seen = ""
repeat = None
for char in text:
    if char in seen:    
        repeat = char
        break
    seen += char       

if repeat:
    print("First repeating character:", repeat)
else:
    print("No repeating characters found.")
```

```python
# 32.Task: Replace Character in String 
# Write a program that takes a string input from the user, then asks for a character 
# to replace and the character to replace it with. The program should output the 
# modified string where all occurrences of the specified character are replaced by 
# the replacement character. 

text = input("Enter a string: ")
old_char = input("Enter the character you want to replace: ")
new_char = input("Enter the new character: ")
modified_text = ""
for ch in text:
    if ch == old_char:
        modified_text += new_char  
    else:
        modified_text += ch        
print("Modified string:", modified_text)

```

```python
# 33.Replace Spaces with Underscores Replace all spaces in a string with underscores (_) 
text = input("Enter a string: ")
modified_text = ""
for ch in text:
    if ch == " ":
        modified_text += "_"  
    else:
        modified_text += ch    
print("Modified string:", modified_text)

```
