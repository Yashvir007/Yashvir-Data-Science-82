```python
#1. Filter vowels and consonants from the string "How are you sir". 
s = "How are you sir"
vowels = "aeiouAEIOU"
vowel_list = [ch for ch in s if ch in vowels]
consonant_list = [ch for ch in s if ch.isalpha() and ch not in vowels]
print("Vowels:", vowel_list)
print("Consonants:", consonant_list)
```

```python
#2. Count vowels and consonants in the string "How are you sir".
s = "How are you sir"
vowels = "aeiouAEIOU"

v = 0
c = 0

for ch in s:
    if ch.isalpha():
        if ch in vowels:
            v += 1
        else:
            c += 1

print("Vowels:", v)
print("Consonants:", c)
```

```python
#12. Check if the string is a palindrome (e.g., "madam" → Palindrome, "hello" → Not palindrome). 

def is_palindrome(str_in):
   s_clean = str_in.replace(' ', '').lower()
   return s_clean == s_clean[::-1]
palindrome_test1 = "madam"
palindrome_test2 = "hello"
is_palindrome_1 = is_palindrome(palindrome_test1)
is_palindrome_2 = is_palindrome(palindrome_test2)
print(is_palindrome_1)
print(is_palindrome_2)
```

```python
#33. Find all occurrences of a substring in "This is Python and Python is fun" → Substring "Python".
text = "This is Python and Python is fun"
substring = "Python"
occurrences = []
start = 0
while True:
start = text.find(substring, start)
if start == -1:
break
occurrences.append(start)
start += 1
print(occurrences)
```
